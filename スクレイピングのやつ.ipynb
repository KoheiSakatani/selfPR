```
#https://newtechnologylifestyle.net/keras%E3%81%AEcnn%E3%82%92%E4%BD%BF%E7%94%A8%E3%81%97%E3%81%A6%E3%82%AA%E3%83%AA%E3%82%B8%E3%83%8A%E3%83%AB%E7%94%BB%E5%83%8F%E3%81%A7%E7%94%BB%E5%83%8F%E8%AA%8D%E8%AD%98%E3%82%92%E8%A1%8C%E3%81%A3/
#を大きく参照

#google Colaboratryでのファイル操作はgdriveを/contentにマウントする必要あり
#これはhttps://qiita.com/uni-3/items/201aaa2708260cc790b8 ,  http://www.bokupy.com/detail/53を参照

import keras
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.layers.core import Dense, Dropout, Activation, Flatten
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image
import glob

#分類するフォルダの名前を格納
folder = ["test_image_shiba","test_image_tosa","test_image_hokkai","test_image_kisyu"]

#50×50にリサイズ→画像によっては引き伸ばしになるため、16×16に変更してみる
image_size = 50

#画像データ,正解ラベルの配列
X = []
Y = []

#画像と正解ラベルの格納
for index, name in enumerate(folder):
  dir = '/content/gdrive/My Drive/test_image/' + name
  files = glob.glob(dir + "/*.jpeg") #拡張子に気を付けて
  for i, file in enumerate(files):
        image = Image.open(file) #フォルダを開く
        image = image.convert("RGB") #RGB値を変換
        image = image.resize((image_size, image_size)) #画像のリサイズ
        data = np.asarray(image) #画像をデータに変換(日本語がおかしいけど伝わるやろ)
        X.append(data)
        Y.append(index)

#扱いやすいようにnumpyに変換
X = np.array(X)
Y = np.array(Y)
print("X")
print(X)
print("X.shape")
print(X.shape)
#3次元を4次元に
#X = np.expand_dims(X, axis=0)

#データの正規化
X = X.astype('float32')
X = X / 255.0 #RGB値は0〜255なので255で割る

# 正解ラベルの形式を変換
Y = np_utils.to_categorical(Y, 4)

# 学習用データとテストデータ
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, shuffle = True)

print("X_train.shape")
print(X_train.shape)

# CNNを構築
model = Sequential()
 
model.add(Conv2D(32, (3, 3), padding='same',input_shape=X_train.shape[1:]))
model.add(Activation('relu'))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

#model.add(Conv2D(128, (3, 3), padding='same'))
#model.add(Activation('relu'))
#model.add(Conv2D(128, (3, 3)))
#model.add(Activation('relu'))
#model.add(MaxPooling2D(pool_size=(2, 2)))
#model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(4))
model.add(Activation('softmax'))
#転移学習を実施(最初の2層を固定)
for layer in model.layers[:2]:
  layer.trainable = False

#モデルの表示
model.summary()

# コンパイル
model.compile(loss='categorical_crossentropy',optimizer='SGD',metrics=['accuracy'])

#訓練
history = model.fit(X_train, y_train, epochs=400)

#評価 & 評価結果出力
print(model.evaluate(X_test, y_test))
```
